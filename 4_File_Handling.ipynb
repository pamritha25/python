# **With statement**

It simplifies  management of common resorces like file streams,the above implementations theres is need to call file.close()
with statement itself ensures proper acquisition and relase of resources.

```python
L = ["This is delhi\n", "this is paris\n", "This is london\n"]
with open("/content/python amru.txt","w")as data1:
    data1.write("hello\n")
    data1.writelines(L)

with open("/content/python amru.txt","a")as data1:
    data1.write("today")

with open("/content/python amru.txt","r+")as data1:
    print(data1.read())
```
